# Notebook 1: Data Validation & Cleaning

Objective: Ensure dataset is clean, consistent, and reliable.

```python
import pandas as pd
import numpy as np

# Load raw data
df = pd.read_csv('../data/raw/restaurant_master.csv')
print('Data loaded. Shape:', df.shape)
df.head()
```

```python
# Validate schema and data types
print('Data types:')
df.dtypes

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])
print('Date converted.')
```

```python
# Check missing values
print('Missing values:')
df.isnull().sum()

# Handle missing values (e.g., fill with median)
df.fillna(df.median(numeric_only=True), inplace=True)
print('Missing values handled.')
```

```python
# Validate revenue = customers × avg_order_value
df['calculated_revenue'] = df['customers'] * df['avg_order_value']
revenue_check = np.isclose(df['revenue'], df['calculated_revenue'], rtol=0.01)
print('Revenue validation passed:', revenue_check.sum() / len(revenue_check))

# Validate profit = revenue – costs
df['calculated_profit'] = df['revenue'] - df['costs']
profit_check = np.isclose(df['profit'], df['calculated_profit'], rtol=0.01)
print('Profit validation passed:', profit_check.sum() / len(profit_check))
```

```python
# Detect unrealistic values
print('Unrealistic values:')
print('Negative customers:', (df['customers'] < 0).sum())
print('Negative revenue:', (df['revenue'] < 0).sum())
print('Extreme revenue (>1e6):', (df['revenue'] > 1e6).sum())

# Cap extreme values
df['revenue'] = np.clip(df['revenue'], 0, 1e6)
df['customers'] = np.clip(df['customers'], 0, None)
print('Unrealistic values capped.')
```

```python
# Save cleaned dataset
df.to_csv('../data/processed/cleaned_restaurant_data.csv', index=False)
print('Cleaned data saved to processed folder.')
```
